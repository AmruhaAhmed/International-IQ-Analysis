```python
import pandas as pd
import io

df = pd.read_csv(io.BytesIO(uploaded['International_IQ.csv']))
print(df)
```

Exploratory Data Analysis

```python
#checking the names of columns
df.columns
```

```python
#checking for any NULL values
df.isnull().sum()
```

```python
#only one country value is NULL, hence we need to drop the record where country is NULL
df=df.dropna();
```

```python
#checking if null values are still present in the dataset or not
df.isnull().sum()
```

```python
df.rename(columns={'Average IQ':'Average_IQ'},inplace=True)
```

```python
df.columns
```

Data Visualization

```python
#importing matplotlib and seaborn libraries for visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=df.sort_values(by=['Average_IQ'],ascending=False)
```

```python
sns.barplot(x='Average_IQ',y='Country',data=df.head(15),palette='magma')
plt.title("Countries with Highest Average IQ")
```

```python
df=df.sort_values(by=['Participants'],ascending=False)
```

```python
sns.barplot(x='Participants',y='Country',data=df.head(),palette='Spectral')
plt.title("Top 5 countries with highest number of Participants")
```

```python
sns.jointplot(y='Participants',x='Average_IQ',data=df,palette='RdPu',kind="hex")
plt.title("Jointplot of AVerage_IQ vs Participants")
```

```python

sns.kdeplot(x='Participants',data=df,shade=True,color='Purple')
plt.title("KDE Plot of Participants")
```

```python
sns.kdeplot(x='Average_IQ',data=df,shade=True,color='Pink')
plt.title("KDE Plot of Average IQ ")
```
